## Adding seeds, and features  and creating game data

```python
def clean_seed(seed):
    name = re.sub(r'[a-zA-Z]', '', seed)  # Remove numbers, periods, and semicolons
    return name.strip()

```

```python
results_path = "data/MNCAATourneyCompactResults.csv"
seeds_path = "data/MNCAATourneySeeds.csv"

seed_df = pd.read_csv(seeds_path)

results_df = pd.read_csv(results_path)

results_df = results_df[results_df["Season"] > 2010]


seed_df["Seed"]  = seed_df["Seed"].apply(clean_seed)
```

```python
all_df =  merged_df.merge(seed_df, on=["TeamID", "Season"])

all_df = all_df.drop(columns=['AdjustO Rank','AdjustD Rank','SOS OppO Rank', 'SOS OppD Rank'])

all_df["Seed"] = all_df["Seed"].astype(int)


# all_df
```

```python
all_df["Conference"].unique()
```

Add Power 5

```python
power_5_conferences = [
    "ACC", "B10", "B12", "P12", "SEC", "BE"
]  # Adjust based on your dataset

# Function to check if a team is in Power 5
def is_power5(conference):
    return 1 if conference in power_5_conferences else 0

# Add Power 5 indicator to all_df
all_df["Power5"] = all_df["Conference"].apply(is_power5)
```

Make winning percentage

```python
all_df[['Wins', 'Losses']] = df['W-L'].str.split('-', expand=True).astype(int)

    # Compute Win Percentage: W / (W + L)
all_df['Win_Percentage'] = all_df['Wins'] / (all_df['Wins'] + all_df['Losses'])

all_df = all_df.drop(columns=["Wins", "Losses"])
```

### Create games from the data. 

```python

# Initialize an empty list to store processed game data
processed_games = []

# Iterate through each game in results_df
for index, row in results_df.iterrows():
    season = row["Season"]
    team_a_id = row["WTeamID"]  # Default: Winner is Team A
    team_b_id = row["LTeamID"]  # Default: Loser is Team B
    swap_flag = np.random.randint(2)  # Randomly decide whether to swap teams

    # Swap teams if `swap_flag == 1`
    if swap_flag == 1:
        team_a_id, team_b_id = team_b_id, team_a_id

    # Get team features from all_df
    # print(team_a_id, season)
    team_a_features = all_df[(all_df["Season"] == season) & (all_df["TeamID"] == team_a_id)]
    team_b_features = all_df[(all_df["Season"] == season) & (all_df["TeamID"] == team_b_id)]
    # print(team_a_features , team_a_id)


    if team_a_features.empty or team_b_features.empty:
        print(f"Skipping game {index}: Missing data for TeamA {team_a_id} or TeamB {team_b_id}")

        print(team_a_features.empty, team_a_id, team_b_features.empty, team_b_id)
        continue

    # Extract numerical features
    feature_columns = ["Rank","Win_Percentage", "AdjustO", "AdjustD", "AdjustT", "Luck", "SOS OppO", "SOS OppD", "Seed", "Power5"]
    team_a_features = team_a_features.iloc[0][feature_columns]
    team_b_features = team_b_features.iloc[0][feature_columns]

    # Compute feature differences
    feature_diffs = team_a_features - team_b_features

    # Assign binary target
    win_label = 1 if swap_flag == 0 else 0  # If swapped, flip the 
    ### essentially the default is that team A won and that categroy is the winner.

    # Store processed row
    processed_games.append({
        "Season": season,
        "TeamA_ID": team_a_id,
        "TeamB_ID": team_b_id,
        **feature_diffs.to_dict(),
        "Win": win_label
    })

# Convert list to DataFrame
final_games_df = pd.DataFrame(processed_games)

```

```python
import itertools

# List of teams in 2025 dataset
teams = df25["TeamID"].unique()

# Feature columns to use
feature_columns = ["Rank", "Win_Percentage", "AdjustO", "AdjustD", "AdjustT", "Luck", "SOS OppO", "SOS OppD", "Seed", "Power5"]

# Initialize an empty list to store matchups
processed_matchups = []

# Iterate through each possible matchup (every team vs. every other team)
for team_a_id, team_b_id in itertools.combinations(teams, 2):
    # Get team stats
    team_a_features = df25[df25["TeamID"] == team_a_id]
    team_b_features = df25[df25["TeamID"] == team_b_id]

    # Ensure both teams exist in the dataset
    if team_a_features.empty or team_b_features.empty:
        print(f"Skipping matchup: {team_a_id} vs. {team_b_id} (Missing data)")
        continue

    # Extract numerical features
    team_a_features = team_a_features.iloc[0][feature_columns]
    team_b_features = team_b_features.iloc[0][feature_columns]

    # Compute feature differences
    feature_diffs = team_a_features - team_b_features

    # Store matchup
    processed_matchups.append({
        "TeamA_ID": team_a_id,
        "TeamB_ID": team_b_id,
        **feature_diffs.to_dict()
    })

# Convert to DataFrame
matchups_df = pd.DataFrame(processed_matchups)

```
